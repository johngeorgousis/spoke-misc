## Product Mapping

https://spokelondon.eu.looker.com/explore/spokelondon/yotpo_upload?qid=7mjwKLVkZw8yOhj8UGsEEf&toggle=fil

```python
product_mapping = pd.read_csv('looker_product_mapping.csv', usecols =
                              ['Yotpo Upload Product Title',
                               'Line Items Title',
                               'Line Items Product Category',
                               'Line Items Product Type',
                               ])

product_mapping.rename(columns={'Yotpo Upload Product Title': 'product_title',
                                'Line Items Title': 'line items colour',
                                'Line Items Product Category': 'line items product category',
                                'Line Items Product Type': 'line items product type'}, inplace=True)
#product_mapping['product_id']=product_mapping['product_id'].astype(int)
product_mapping = product_mapping[['product_title','line items colour','line items product category','line items product type']]
product_mapping
```

## Join Mapping to Merged Yotpo

```python
product_merge = merge.merge(product_mapping, on=['product_title'],how='left')
#product_merge.to_csv('product_merge.csv')
product_merge
```

```python
product_merge = product_merge.drop_duplicates(subset=['id'])
product_merge = product_merge.sort_values(by="date")
product_merge.shape
product_merge['incremental_id'] = range(0, len(product_merge))
product_merge.to_csv('full_YOTPO_upload_18032022.csv',index=False)
#product_merge['modifed_at'] = pd.Timestamp.now(tz=None)
product_merge
```
